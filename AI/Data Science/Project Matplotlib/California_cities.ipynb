### Data visualization of cities in the state of california in the US (trực quan hóa dữ liệu các thành phố ở bang california ở mỹ)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities=pd.read_csv("california_cities.csv")
cities.head()
cities['area_total_km2'].sort_values(ascending=False)
```

```python
# Extract latitude and longitude data (Trích xuất dữ liệu vĩ độ và kinh độ)
lat,lon = cities['latd'],cities['longd']

# Extract population and area data (Trích xuất dữ liệu dân số và diện tích)
population, area = cities['population_total'],cities['area_total_km2']

# Set the plot style and size (Đặt kiểu và kích thước)
plt.style.use('seaborn')
plt.figure(figsize=(8,6))

# Create a scatter plot of longitude and latitude, with marker size and color determined by city area and population, respectively (Tạo một biểu đồ phân tán kinh độ và vĩ độ, với kích thước và màu sắc của điểm đánh dấu được xác định tương ứng theo diện tích thành phố và dân số)
plt.scatter(lon,lat,
            c = np.log10(population),cmap="viridis",linewidths=0,alpha=0.5,
            s= area)

# Set the axis limits to be equal, and label the x and y axes (Đặt các giới hạn trục bằng nhau và gắn nhãn các trục x và y)
plt.axis('equal')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add a colorbar indicating the logarithm of city population, with a label (Thêm một thanh màu biểu thị logarit của dân số thành phố, có nhãn)
plt.colorbar(label = 'log$_{10}$(population)')
plt.clim(3,7)

# Create a legend for city sizes, with markers for each of four different area ranges(Tạo chú giải cho quy mô thành phố, với các điểm đánh dấu cho từng phạm vi trong số bốn phạm vi khu vực khác nhau)
area_range = [50,100,300,500]
for area in area_range:
    plt.scatter([],[],s =area,label = str(area) + "km$^2$" ,alpha=0.5)

plt.legend(labelspacing =1,title = "City Area")

# Add a title to the plot (Thêm tiêu đề)
plt.title('California Cities: Population and Area Distribution')
```
