## Introduction to Matplotlib (Giới thiệu về Matplotlib)

Matplotlib là một thư viện đồ họa trong Python được sử dụng để tạo ra các biểu đồ, đồ thị và hình ảnh tương tác trong các ứng dụng phân tích dữ liệu và khoa học dữ liệu. Nó cho phép người dùng dễ dàng tạo ra các biểu đồ chất lượng cao bằng cách cung cấp một loạt các tính năng và tùy chọn để tùy chỉnh màu sắc, font chữ, kích thước và kiểu dáng của đồ thị.

Matplotlib cung cấp cho người dùng các giao diện trực quan để tạo ra các biểu đồ và đồ thị, bao gồm các giao diện đồ họa và đối tượng hướng đối tượng. Nó cũng cho phép người dùng lưu các đồ thị của họ dưới dạng các tệp hình ảnh như PNG, PDF, SVG và nhiều định dạng hình ảnh khác.

Matplotlib là một trong những thư viện đồ họa phổ biến nhất và được sử dụng rộng rãi trong cộng đồng Python. Nó được phát triển bởi John D. Hunter vào năm 2003 và hiện được duy trì bởi một nhóm các nhà khoa học dữ liệu và lập trình viên.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

plt.plot() là một hàm trong thư viện Matplotlib của Python, được sử dụng để tạo đồ thị đường (line plot) từ các tập dữ liệu đã cho. Hàm này nhận các đối số đầu vào là các mảng hoặc danh sách các giá trị số, tương ứng với trục x và trục y của đồ thị, và cung cấp các tùy chọn để định dạng và trình bày đồ thị.

Ví dụ, để vẽ đồ thị đường của một tập dữ liệu x và y, ta có thể sử dụng lệnh plt.plot(x, y), trong đó x và y là các mảng chứa các giá trị tương ứng trên trục x và y. Hàm plt.plot() có thể được sử dụng để thêm nhiều đường trên cùng một đồ thị bằng cách gọi nó nhiều lần với các cặp dữ liệu khác nhau.

Các tùy chọn định dạng và trình bày đồ thị bao gồm: màu sắc đường, kiểu đường, độ dày của đường, các nhãn trục, tiêu đề đồ thị, vùng mờ, lưới đồ thị, độ dày của biểu tượng, các giá trị trục, vị trí và định dạng của các nhãn trục, kích thước đồ thị và tỷ lệ khung hình.

Với plt.plot(), người dùng có thể tạo đồ thị đường đơn giản và dễ đọc từ các tập dữ liệu số, thường được sử dụng trong phân tích dữ liệu và khoa học dữ liệu.

## Two ways to graph Matplotlib (Hai cách để vẽ đồ thị Matplotlib )
### Pyplot API vs Object-Oriented (OO) API
* Pyplot API -> Quickly
* OO API -> Advanced



* #### Pyplot API

```python
x = np.linspace(0,10,1000)
x
```

```python
np.sin(x)
```

```python
# x = np.linspace(0,10,10)
# color để đặt màu sắc cho đường đồ thị
# linestyle để đặt phong cách cho đường đồ thị
# label đặt nhẵn cho đường đồ thị
plt.plot(x,np.sin(x),color = 'blue',linestyle ='dashed',label = 'sin(x)'); 
plt.plot(x, np.cos(x),color = 'red',label = 'cos(x)');

plt.title('Sin(X) and Cos(X)')

plt.xlabel('X'); # đặt nhãn cho trục x của đồ thị
plt.ylabel('F(X)'); # đặt nhẵn cho trục y của đồ thị

# plt.xlim([0,4]) : xlim để đặt giới hạn cho trục hoành (x)
# plt.ylim([0,4]) : ylim để đặt giới hạn cho trục tung (y)
# plt.axis([0,4,0,1]) : axis để đặt giới hạn cho trục hoành (x) và (y) .
# Cú pháp : plt.axis([minX,maxX,minY,maxY])
# plt.axis('tight')
plt.axis('equal')

plt.legend();  # thêm chú thích (legend) cho đồ thị. 
#plt.show();
```

* #### Object-Oriented (OO) API

```python
#1. Prepare data    
x = [10,45]
y = [40,0]
x1 = [10,45]
y1 = [0,40]
#2. Setup plot
fig,ax = plt.subplots(figsize=(5,5)) #Figure size = width & height of the plot

#3. Plot the data 
ax.plot(x,y,color = "red",linestyle = "dashed",label = "test");
ax.plot(x1,y1,color = "blue",linestyle = "dashed",label = "none");

ax.set(title="Bản Đồ Hải Tặc",xlabel= "x",ylabel="y");
ax.legend();

    
```

## 2.Most common types of Matplotlib plots (Các loại biểu đồ và đồ thị phổ biến trong Matplitlib)
* line (Biểu đồ đường)
* scatter (Biểu đồ phân tán)

#### Lines (Biểu đồ đường)
Biểu đồ đường (line chart): Biểu đồ đường được sử dụng để thể hiện sự thay đổi của một biến hoặc nhiều biến theo thời gian. Biểu đồ này thường được sử dụng để thể hiện xu hướng, mối tương quan hoặc sự biến động của dữ liệu theo thời gian.

```python
# Create an array
x = np.linspace(0,10,100)
x[:5]
```

```python
plt.style.use("ggplot")
```

```python
## OO API plot a line chart

fig,ax = plt.subplots()
ax.plot(x, x**3,color = "blue")

```

#### Scatter (Biểu đồ phân tán)
Biểu đồ phân tán (scatter plot): Biểu đồ phân tán được sử dụng để thể hiện mối quan hệ giữa hai biến. Mỗi điểm trên biểu đồ đại diện cho một cặp giá trị của hai biến. Biểu đồ phân tán được sử dụng để tìm kiếm mối quan hệ và kiểm tra sự tương quan giữa hai biến.

```python
# Pyplot API
# Prepare New Data
i = np.random.randint(1,100) # random từ 1,100
rng = np.random.RandomState(i) # hạt giống random trong khoảng i
x = rng.randn(100) # in ra 100 giá trị theo hạt giống random
y = rng.randn(100) # in ra 100 giá trị theo hạt giống random
colors = rng.randn(100) # in ra 100 giá trị theo hạt giống random 
sizes = 1000* rng.randn(100) # mỗi phần tử sẽ tăng lên 1000
plt.scatter(x,y,c = colors,s = sizes,cmap="viridis",alpha= 0.3); 
# x và y: là hai mảng numpy chứa giá trị của trục hoành và trục tung, tương ứng với các điểm trên biểu đồ.

# c: là một mảng numpy chứa các giá trị màu sắc của các điểm trên biểu đồ. Mỗi giá trị trong mảng tương ứng với một điểm trên biểu đồ.

# s: là một mảng numpy chứa kích thước của các điểm trên biểu đồ. Mỗi giá trị trong mảng tương ứng với một điểm trên biểu đồ.

# cmap: là một chuỗi xác định bản đồ màu (color map) được sử dụng để chuyển đổi giá trị màu trong mảng c thành các giá trị màu trên biểu đồ.

# alpha: là một số thực từ 0 đến 1, xác định độ trong suốt của các điểm trên biểu đồ.
    
```

```python
# OO API

fig,ax = plt.subplots()
img = ax.scatter(x,y,c = colors,s = sizes,cmap="viridis",alpha= 0.3);
fig.colorbar(img)
```

#### Bar (Biểu đồ cột)
* Vertical (Thẳng đứng)
* Horizontal (Nằm ngang)

Biểu đồ cột (bar chart): Biểu đồ cột được sử dụng để so sánh giá trị của các biến trên một tập hợp các danh mục hoặc nhóm. Biểu đồ này thường được sử dụng để so sánh giá trị của các biến trên các nhóm khác nhau.

```python
# Vertical(Thẳng đứng)
# Prepare Data From A Python Dictionary
soft_drink_prices =  {"Coke":10,
                      "Pessi":12,
                      "Sprite":8}
fig,ax = plt.subplots()
ax.bar(soft_drink_prices.keys(),soft_drink_prices.values(),color = "blue");
ax.set(title="Bach Hoa Xanh's",ylabel= "Prices ($)");
```

```python
# Horizontal(Nằm ngang)
fig,ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()),list(soft_drink_prices.values()));
ax.set(title="Bach Hoa Xanh's",xlabel= "Prices ($)");
```

#### Histogram (Biểu đồ thể hiện tần suất dạng cột)
Biểu đồ thể hiện tần suất dạng cột (histogram): Biểu đồ histogram được sử dụng để thể hiện phân phối tần suất của một biến. Biểu đồ này chia các giá trị của biến thành các khoảng dữ liệu (bins) và đếm số lượng giá trị của biến trong mỗi khoảng đó. Biểu đồ histogram thường được sử dụng để khám phá phân phối của dữ liệu và kiểm tra tính đối xứng của phân phối.

```python
# Prepare Data
np.random.seed(42)
students_height = np.random.normal(170 ,10, 250)
students_height[:5],len(students_height)
```

```python
# Pyplot API
plt.hist(students_height,color="blue",bins=20);
plt.title("Students Height");
plt.xlabel("Đơn vị : m");
plt.ylabel("Số lượng");
```

```python
# OO API
fig,ax = plt.subplots()
ax.hist(students_height,bins=50);
ax.set(title="Students Height 2",xlabel="đơn vị : m",ylabel="Số Lượng");
```

#### Subplots

```python
# Option 1
fig, ((ax1,ax2),(ax3,ax4))= plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize =(10,5))
# Prepare new data 
x = np.linspace(0,10,100);
ax1.plot(x,x/2);
ax1.plot(x,x*2);

# scatter
ax2.scatter(np.random.random(10),np.random.random(10));
# bar
ax3.bar(soft_drink_prices.keys(),soft_drink_prices.values());
# hist
ax4.hist(students_height,bins=30);
```
